## 3.1.2

### Importing the stats module from scipy

```python
from scipy import stats
```

### Running a 1 sample T-test to see the probability of the population mean being 0 for VIQ, it seems to be quite unlikely

```python
stats.ttest_1samp(data['VIQ'], 0)   
```

### Running a 2 sample T-test to see if there is a significant difference between amle and female VIQ's, there doesn't seem to be

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)   
```

### Running a 2 sample T-test to see if FSIQ and PIQ are statistically significantly different, they are not

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])   
```

### Rerunning the above T-test, but removing inter-subject variability, ah much better

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])   
```

### Runnign a similar test, but on a model that assumes a less gaussian model of the data

```python
stats.wilcoxon(data['FSIQ'], data['PIQ'])   
```

### Importing numpy and generating random values in an array and then creating a model from said data

```python
import numpy as np
x = np.linspace(-5, 5, 20)
np.random.seed(1)
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)
data = pandas.DataFrame({'x': x, 'y': y})
```

### Importing the OLS model module and running it for the data

```python
from statsmodels.formula.api import ols
model = ols("y ~ x", data).fit()
```

```python
print(model.summary())  
```

### Creating an OLS model for the brain sizes data VIQs

```python
data = pandas.read_csv('brain_size.csv', sep=';', na_values=".")
model = ols("VIQ ~ Gender + 1", data).fit()
print(model.summary())  
```

### Displaying IQ scores by type of IQ test

```python
data_fisq = pandas.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pandas.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pandas.concat((data_fisq, data_piq))
print(data_long)  
```

### Creating an OLS model of the iq types from above

```python
model = ols("iq ~ type", data_long).fit()
print(model.summary())  
```

### Creating an OLS model from the iris data

```python
data = pandas.read_csv('iris.csv')
model = ols('sepal_width ~ name + petal_length', data).fit()
print(model.summary())
```

### running an f test, another non-gaussian distribution, on the above variables

```python
print(model.f_test([0, 1, -1, 0]))
```

### importing the education data, had to modify urllib to execute correctly

```python
import pandas
import urllib.request

if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')

# Give names to the columns
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None)
data.columns = short_names

# Log-transform the wages, because they typically are increased with
# multiplicative factors
import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```

### Creating Graphs of the education wage data via seaborn

```python
import seaborn
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg')  
```

### Creating a different set of graphs of the education wage data, but color coding on the basis of sex

```python
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg', hue='SEX') 
```

### Creating a linear regression of wage vs education with seaborn

```python
seaborn.lmplot(y='WAGE', x='EDUCATION', data=data)  
```
